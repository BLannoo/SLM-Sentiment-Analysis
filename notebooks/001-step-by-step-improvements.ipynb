```python
from time import time
from llama_cpp import Llama
```

```python
from transformers import set_seed

set_seed(42)

llm_qwen = Llama.from_pretrained(
    repo_id="Qwen/Qwen2-1.5B-Instruct-GGUF",
    # See the all quantization levels here:
    # https://huggingface.co/Qwen/Qwen2-1.5B-Instruct-GGUF/tree/main
    # It's better to start from *q5_k_m, and if it's too bad - continue with q6_k and q8_0.
    filename="qwen2-1_5b-instruct-q5_k_m.gguf",
    verbose=False,
    n_ctx=32768,
)
```

```python
def run(prompt, reviews):
    for review, actual_sentiment in reviews:
        start_time = time()
        full_prompt = prompt.format(review=review)
        response = llm_qwen.create_chat_completion(
            temperature=0.01,
            messages=[
                {
                    "role": "system",
                    "content": "You are a helpful AI assistant.",
                },
                {
                    "role": "user",
                    "content": full_prompt,
                },
            ]
        )
        print(f"===== Review ===== (with actual sentiment: {actual_sentiment})")
        print(review)
        print(f"===== Response ===== (with actual sentiment: {actual_sentiment})")
        print(response["choices"][0]["message"]["content"])
        print(f"===== Timing ===== (with actual sentiment: {actual_sentiment})")
        print(time() - start_time)
```
