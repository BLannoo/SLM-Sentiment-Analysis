# **Aside On:** Confidence Intervals for Proportions: Wald and Wilson Intervals

When estimating a population proportion $p$ from a sample proportion $\hat{p}$, it's essential to calculate confidence intervals to understand the range in which the true proportion is likely to lie. Two common methods for calculating these intervals are the **Wald Interval** and the **Wilson Interval**.

## Wald Interval

The **Wald Interval** for a proportion is based on the normal approximation to the binomial distribution. It is defined as:

$$
\hat{p} \pm z \cdot \text{SE}(\hat{p})
$$

where:

- $\hat{p}$ is the sample proportion,
- $n$ is the sample size,
- $z$ is the z-score corresponding to the desired confidence level (e.g., 1.96 for a 95% confidence level),
- $\text{SE}(\hat{p})$ is the standard error of $\hat{p}$, given by:

$$
\text{SE}(\hat{p}) = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
$$

### Wald Criteria

The **Wald Criteria** determine when the Wald Interval is reliable. The criteria are:

- $n \cdot \hat{p} \geq 5$
- $n \cdot (1 - \hat{p}) \geq 5$

If both conditions are met, the normal approximation is considered reasonable, and the Wald interval is likely to provide a good estimate.

## Wilson Interval

The **Wilson Interval** is a more accurate alternative to the Wald Interval, especially for small sample sizes or when $\hat{p}$ is near 0 or 1. The Wilson interval is given by:

$$
\frac{\hat{p} + \frac{z^2}{2n}}{1 + \frac{z^2}{n}} \pm \frac{z}{1 + \frac{z^2}{n}} \sqrt{\frac{\hat{p}(1 - \hat{p})}{n} + \frac{z^2}{4n^2}}
$$

where:

- $\hat{p}$ is the sample proportion,
- $n$ is the sample size,
- $z$ is the z-score corresponding to the desired confidence level.

The **Wilson Interval** adjusts for skewness in the distribution of $\hat{p}$ and provides more accurate confidence intervals, particularly when the sample size is small or when the proportion is extreme.

## Summary

- **Wald Interval**: Easy to compute but less reliable for small sample sizes or extreme proportions.
- **Wald Criteria**: Guidelines to ensure the normal approximation is valid for the Wald Interval.
- **Wilson Interval**: A more accurate interval for proportions, particularly when the sample size is small or the proportion is close to 0 or 1.
